```python
import yaml

data = { 'train': '/content/drive/MyDrive/dataset/train/images',
'val': '/content/drive/MyDrive/dataset/train/images',
'test': '/content/drive/MyDrive/dataset/train/images',
'nc': 1,
'names': ['building']}

with open('/content/drive/MyDrive/dataset/data.yaml', 'w') as file:
    yaml.dump(data, file)

with open('/content/drive/MyDrive/dataset/data.yaml', 'r') as file:
    yaml=yaml.safe_load(file)
    display(yaml)
```

```python
from ultralytics import YOLO

model = YOLO('yolov8n.pt')
model.train(data='/content/drive/MyDrive/dataset/data.yaml', epochs=100)
```
