```python
from ultralytics import YOLO
import cv2
model = YOLO("./runs/detect/train/weights/best.pt")
```

```python
img_path = 'img/1ZAIQ333LB_1.jpg'
source = cv2.imread(img_path)
results = model.predict('img/1ZAIQ333LB_1.jpg')
```

```python
from ultralytics.utils.plotting import Annotator

# Extract results
annotator = Annotator(source, example=model.names)

for box in results[0].boxes.xyxy.cpu():
    width, height, area = annotator.get_bbox_dimension(box)
    print("Bounding Box Width {}, Height {}, Area {}".format(
        width.item(), height.item(), area.item()))
```

```python
import cv2
import numpy as np
from ultralytics import YOLO

# YOLO  
model = YOLO("./runs/detect/train/weights/best.pt")
image_path = 'img/1ZAIQ333LB_1.jpg'

#  
img = cv2.imread(image_path)

#   
results = model(img)

#  
detections = results[0].boxes.data.cpu().numpy()

#  
mask = np.zeros(img.shape[:2], dtype=np.uint8)

for *box, conf, cls in detections:
    if cls == 0:  # class 0 building
        x1, y1, x2, y2 = map(int, box)
        mask[y1:y2, x1:x2] = 255
    elif cls == 1:  # class 1 sign
        x1, y1, x2, y2 = map(int, box)
        mask[y1:y2, x1:x2] = 255

# inpainting 
inpainted_img = cv2.inpaint(img, mask, inpaintRadius=3, flags=cv2.INPAINT_TELEA)

#  
cv2.imwrite('inpainted_image.jpg', inpainted_img)
```
